# Challenge 2 - Hypothesis Tests

In this section of the lab, we will test whether the hourly wage of all hourly workers is significantly different from $30/hr. Import the correct one sample test function from scipy and perform the hypothesis test for a 95% two sided confidence interval.

```python
# Your code here:
data1 = data[data['Hourly Rate'].notna()]
data1['Hourly Rate'] = data1['Hourly Rate'].str.replace('$', '')
data1['Hourly Rate'] = pd.to_numeric(data1['Hourly Rate'])

```

```python
degrees_freedom = len(data1['Hourly Rate'])
degrees_freedom
```

```python
sample_mean = np.mean(data1['Hourly Rate'])
sample_mean
```

```python
sample_standard_error = scipy.stats.sem(data1['Hourly Rate'])
```

```python
import scipy.stats

confidence_level = 0.95

confidence_interval = scipy.stats.t.interval(confidence_level, 
                                             degrees_freedom, 
                                             sample_mean, 
                                             sample_standard_error)
```

```python
print( 'confidence interval is ', confidence_interval, '.' )
```

```python
# Two-sided test:
# Null hypothesis or H0: mean value = 30
# Alternative hyp or H1: mean value != 30

from scipy.stats import ttest_1samp

stat, pval = ttest_1samp(data1['Hourly Rate'], 30)
```

```python
print('stat is  ', stat)
print('pvalue for the two-tailed test is ', pval)
```

We are also curious about salaries in the police force. The chief of police in Chicago claimed in a press briefing that salaries this year are higher than last year's mean of $86000/year a year for all salaried employees. Test this one sided hypothesis using a 95% confidence interval.

Hint: A one tailed test has a p-value that is half of the two tailed p-value. If our hypothesis is greater than, then to reject, the test statistic must also be positive.

```python
# Your code here:
data2 = data[data['Department']== 'POLICE']
data2 = data2[data2['Annual Salary'].notna()]
data2['Annual Salary'] = data2['Annual Salary'].str.replace('$', '')
data2['Annual Salary'] = pd.to_numeric(data2['Annual Salary'])
```

```python
confidence_level = 0.95
degrees_freedom = len(data2['Annual Salary']) - 1  
sample_mean = np.mean(data2['Annual Salary'])

sample_standard_error = scipy.stats.sem(data2['Annual Salary'])

confidence_interval = scipy.stats.t.interval(confidence_level, 
                                             degrees_freedom, 
                                             sample_mean, 
                                             sample_standard_error)
```

```python
print( 'confidence interval is ', confidence_interval, '.' )
```

```python
# One-sided test:
# Null hypothesis or H0: mean value > 86000
# Alternative hyp or H1: mean value < 86000
from scipy.stats import ttest_1samp

stat, pval = ttest_1samp(data2['Annual Salary'], 86000)
```

```python
print('stat is  ', stat)

print('pvalue for the one-tailed test is ', pval/2)
```

```python
#LEFT-one-sided-test: Chief of Police is right because it's a LEFT-sided-Test and the p-value is close to 0.
```

# Challenge 3: To practice - Constructing Confidence Intervals

While testing our hypothesis is a great way to gather empirical evidence for accepting or rejecting the hypothesis, another way to gather evidence is by creating a confidence interval. A confidence interval gives us information about the true mean of the population. So for a 95% confidence interval, we are 95% sure that the mean of the population is within the confidence interval. 
).

To read more about confidence intervals, click [here](https://en.wikipedia.org/wiki/Confidence_interval).


In the cell below, we will construct a 95% confidence interval for the mean hourly wage of all hourly workers. 

The confidence interval is computed in SciPy using the `t.interval` function. You can read more about this function [here](https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.t.html).

To compute the confidence interval of the hourly wage, use the 0.95 for the confidence level, number of rows - 1 for degrees of freedom, the mean of the sample for the location parameter and the standard error for the scale. The standard error can be computed using [this](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.sem.html) function in SciPy.

```python
# Your code here:

confidence_level = 0.95
degrees_freedom = len(data3['Hourly Rate']) - 1  
sample_mean = np.mean(data3['Hourly Rate'])

sample_standard_error = scipy.stats.sem(data3['Hourly Rate'])

confidence_interval = scipy.stats.t.interval(confidence_level, 
                                             degrees_freedom, 
                                             sample_mean, 
                                             sample_standard_error)
```

```python
print( 'confidence interval is ', confidence_interval, '.' )
```
